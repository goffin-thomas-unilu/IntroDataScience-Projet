```python
# D
```

```python
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score
```

```python
input = [('sc1', StandardScaler()), ('polyn', PolynomialFeatures(degree = 2)), ('lm',LinearRegression())]
pipe = Pipeline(input)
pipe.fit(X,y)
yhatPipe = pipe.predict(X)
Pipe_r2 = r2_score(y,yhatPipe)
print(f"Pipeline r²: {Pipe_r2:.3f}")
yhatPipe
```

```python
print("The pipeline (3.D, R² = 0.582) slightly outperforms polynomial regression (3.C, R² = 0.58, MAE = 11.42) and significantly beats linear regression (3.B, R² = 0.39, MAE = 13.711) for predicting thalach. Its predictions (101.97–196.55) match 3.C’s variability, capturing complex relationships better than 3.B’s (115.34–182.04).")
```

```python
# E
```

```python
from sklearn.model_selection import train_test_split
```

```python
degrees = range(1, 7)
r2_train_scores = []
r2_test_scores = []

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
print(f"Shape of X_train : {X_train.shape}, X_test : {X_test.shape}")

for degree in degrees:
    pipeline = Pipeline([('scaler', StandardScaler()),('poly', PolynomialFeatures(degree=degree)),('linear', LinearRegression())])
    pipeline.fit(X_train, y_train)
    
    r2_train = r2_score(y_train, pipeline.predict(X_train))
    r2_test = r2_score(y_test, pipeline.predict(X_test))
    
    r2_train_scores.append(r2_train)
    r2_test_scores.append(r2_test)
    
    print(f"Degree {degree} : R² Train = {r2_train:.3f}, R² Test = {r2_test:.3f}")

optimal_degree = degrees[np.argmax(r2_test_scores)]
print(f"\n3.E : Optimal polynomial degree : {optimal_degree} (R² Test = {max(r2_test_scores):.3f})")
```

```python
# We will use a Confusion Matrix to compute the accuracy of our model

"""
Confusion matrix provides statistics to evaluate the accuracy of a classification model :
                Predicted positive       Predicted negative
Actual Positive          TP                      FN
Actual Negative          FP                      TN

True Positive (TP) : The number of instances correctly predicted as positive by the model
True Negative (TN) : The number of instances correctly predicted as negative by the model
False Positive (FP) : The number of instances incorrectly predicted as positive by the model (also known as a Type I error)
False Negative (FN) : The number of instances incorrectly predicted as negative by the model (also known as a Type II error)

The precision : the accuracy of the positive predictions    :
    Precision = TP / TP+FP
Recall : also called sensitivity or true positive rate TPR , this is the ratio of positive instances that are correctly detected by the classifier
    Recall = TP / TP+FN

We can combine precision and recall into a single metric called F1 score to compare two classifiers
The F1 score is the harmonic mean of precision and recall
    F1 = TP / TP + (FN+FP)/2
"""
```

```python

# compute precision, recall, f1
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred)
print('precision: ',precision)
print('recall: ',recall)
print('f1: ',f1)
```
